```python
# Import necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Define the URL of the webpage you want to scrape
url="https://www.iplt20.com/auction/2022"
# Send a GET request to the URL
r=requests.get(url)
#print(r)
# Parse the HTML content using BeautifulSoup
soup=BeautifulSoup(r.text,"lxml")
#print(soup)
# Find the table element with the specified class
table=soup.find("table",class_="ih-td-tab auction-tbl")
#print(table)
# Find all the header cells (<th>) in the table
title=table.find_all("th")
# Print the header cells
print(title)
```

```python
# Create an empty list to store the header names
header=[]
# Iterate through the header cells (<th>)
for i in title:
  # Extract the text from the header cell and strip any whitespace
  name=i.text
  # Append the header name to the list
  header.append(name)
# Print the list of header names
print(header)
```

```python
# Create a Pandas DataFrame with the specified header names
df=pd.DataFrame(columns=header)
df
```

```python
import pandas as pd

# Create an empty DataFrame with the specified header names
df = pd.DataFrame(columns=['TEAM', 'FUNDS REMAINING', 'OVERSEAS PLAYERS', 'TOTAL PLAYERS'])
# Find all the table rows in the table
rows=table.find_all("tr")
# Iterate through each row and extract the data
for i in rows[1:]:
    # Extract the team name from the first table cell
    first_td = i.find_all('td')[0].find('div', class_='ih-pt-ic').text.strip()
    # Extract the remaining data from the other table cells
    data = i.find_all('td')[1:]
    row = [tr.text.strip() for tr in data]
    # Insert the team name at the beginning of the row
    row.insert(0, first_td)
    # Append the row to the DataFrame
    df = df.append(pd.DataFrame([row], columns=['TEAM', 'FUNDS REMAINING', 'OVERSEAS PLAYERS', 'TOTAL PLAYERS']), ignore_index=True)

# Print the DataFrame
print(df)
```
